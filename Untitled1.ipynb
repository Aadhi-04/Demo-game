analyzing the data

```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

filename = './3D.fits'
hdul = fits.open(filename)

# Explore available column names
print(hdul[1].columns.names)
```

Extracting the data

```python
data = hdul[1].data

time = data['TIME']
flux = data['PDCSAP_FLUX']  
```

Have to clean the data from noise present 

Importing BoxLeastSquares to see the dip of the exoplanet 

```python
from astropy.timeseries import BoxLeastSquares
```

```python
import matplotlib.pyplot as plt
#Basic mask: remove NaNs and zero/negative flux
mask = np.isfinite(time) & np.isfinite(flux) & (flux > 0.5)
time = time[mask]
flux = flux[mask]
```

Plotting the curve

```python
plt.figure()
plt.plot(time, flux, ".", markersize=1)
plt.xlabel("Time (BJD)")
plt.ylabel("Normalized Flux")
plt.title("Cleaned Light Curve")
plt.savefig("1_cleaned_lightcurve.png")
#plt.close()
```

After chapt gpt and a search online we found the BLS method to detect exoplanets

```python

bls = BoxLeastSquares(time, flux)
periods = np.linspace(0.5, 10, 10000)  # Search between 0.5 and 10 days
duration = 0.1  # Trial transit duration in days

results = bls.power(periods, duration)
# Best-fit parameters
best_index = np.argmax(results.power)
best_period = results.period[best_index]
best_duration = results.duration[best_index]
best_transit_time = results.transit_time[best_index]
best_depth = results.depth[best_index]
```

Plot a peridogram to analyse dips

```python
plt.figure()
plt.plot(results.period, results.power, color='black')
plt.xlabel("Period (days)")
plt.ylabel("BLS Power")
plt.title("BLS Periodogram")
plt.axvline(best_period, color='red', linestyle='--', label=f"Best period = {best_period:.4f} days")
plt.legend()
plt.savefig("2_bls_periodogram.png")
#plt.close()
```
